```python
matfn = 'C:\\Users\\yht\\Desktop\\dataBaseInNum\\feature\\use_direct\\feature.mat'
data = sio.loadmat(matfn)
X = data['feature']
matfn = 'C:\\Users\\yht\\Desktop\\dataBaseInNum\\feature\\use_direct\\label.mat'
data = sio.loadmat(matfn)
y = data['label']
matfn = 'C:\\Users\\yht\\Desktop\\dataBaseInNum\\feature\\use_direct\\label_used.mat'
y_used = sio.loadmat(matfn)
y_used = y_used['label_used']
y1 = y_used[:,0]
y2 = y_used[:,1]
y_multi = np.stack((y1,y2),axis = -1)
```

```python
#标准化
data_standard = preprocessing.StandardScaler().fit(X)
data = data_standard.transform(X)
n_samples, n_features = data.shape
ecg_feature = data[:,:50]
ibi_feature = data[:,50:81]
sc_feature = data[:,81:97]
rsp_feature = data[:,97:]
```

# 根据不同的特征绘制曲线,这部分是非线性的分析，目的是证明情绪识别是有效的

```python
#1、tsne
#http://scikit-learn.org/stable/auto_examples/manifold/plot_lle_digits.html#sphx-glr-auto-examples-manifold-plot-lle-digits-py
data_tsne = TSNE(n_components=2).fit_transform(data)#, perplexity = 5
x_min, x_max = np.min(data_tsne, 0), np.max(data_tsne, 0)
data_tsne = (data_tsne - x_min) / (x_max - x_min)

plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=y1,cmap=plt.cm.Spectral)#)#
plt.show()
```

```python
#2、LDA效果最好
#http://scikit-learn.org/stable/auto_examples/classification/plot_lda.html#sphx-glr-auto-examples-classification-plot-lda-py
#4分类
data_LDA = LinearDiscriminantAnalysis().fit_transform(data,y.ravel())#solver='lsqr', shrinkage='auto'
#plt.scatter(data_LDA[:, 0], data_LDA[:, 1], c = y, cmap=plt.cm.Spectral)
color = {1:'red',2:'blue', 3: 'green',4:'yellow'}
plt.scatter(data_LDA[:17, 0], data_LDA[:17, 1], c = 'red', label = 'peace')
plt.scatter(data_LDA[17:34, 0], data_LDA[17:34, 1], c = 'blue', label = 'happy')
plt.scatter(data_LDA[34:51, 0], data_LDA[34:51, 1], c = 'green', label = 'sadness')
plt.scatter(data_LDA[51:, 0], data_LDA[51:, 1], c = 'yellow', label = 'fear')
plt.legend()
plt.show()
#下面的是用的原数据
data_LDA2 = LinearDiscriminantAnalysis().fit(X,y.ravel())
data_LDA2.transform(X)
```

```python
halfdata = n_samples/2
plt.scatter(data_LDA[:halfdata, 0], data_LDA[:halfdata, 1], marker='x',
            color='blue', alpha=0.7, label='y1')
plt.scatter(data_LDA[halfdata:,0], data_LDA[halfdata:,1], marker='o',
            color='red', alpha=0.7, label='y1')
plt.title('Basic scatter plot')
plt.ylabel('feature1')
plt.xlabel('feature2')
plt.legend()#loc='upper left'
plt.show()
```

```python
#2分类,y1
data_LDA = LinearDiscriminantAnalysis().fit_transform(data,y1)#solver='lsqr', shrinkage='auto'
LDA2 = np.concatenate((data_LDA,y1.reshape(-1,1)),axis=1)
df = pd.DataFrame(LDA2,columns = ['feature','type'])
sns.boxplot(x= 'type', y='feature', data=df)
sns.swarmplot(x= 'type', y='feature', data=df, color=".25")
plt.show()
```

# 下面是对特征级的分析

```python
matfn = 'C:\\Users\\yht\\Desktop\\dataBaseInNum\\feature\\use_direct\\featname.mat'
featname = sio.loadmat(matfn)
featname = featname['featname']
target_names = ['peace','happy','sadness','fear']
```

```python
fname = []
for i in xrange(n_features):
    fname.append(str(featname[0][i][0]))
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 5)
clf = clf.fit(X, y)
```

```python
from IPython.display import Image  
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=fname,  
                         class_names=target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data) 
Image(graph.create_png())
```

```python
clf = tree.DecisionTreeClassifier(max_depth=2)
clf = clf.fit(data_LDA[:,:2], y)

x_min, x_max = data_LDA[:, 0].min() - 1, data_LDA[:, 0].max() + 1
y_min, y_max = data_LDA[:, 1].min() - 1, data_LDA[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(data_LDA[:, 0], data_LDA[:, 1], c = y, alpha=0.8, cmap=plt.cm.Spectral)
plt.show()
```
